```python
import pandas as pd
import numpy as np

# Define the number of samples
num_samples = 3000  

# Generate random pH values (range: 4.5 - 9.0)
pH_values = np.round(np.random.uniform(4.5, 9.0, num_samples), 2)

# Generate random moisture percentage (range: 5% - 50%)
moisture_values = np.round(np.random.uniform(5, 50, num_samples), 2)

# Generate random bearing capacity (kN/m²) (range: 50 - 450)
bearing_capacity_values = np.round(np.random.uniform(50, 450, num_samples), 2)

# Define soil types
soil_types = ["Sandy", "Clay", "Loamy", "Silty", "Peaty"]
soil_type_values = np.random.choice(soil_types, num_samples)

# Create DataFrame
soil_data = pd.DataFrame({
    "pH": pH_values,
    "Moisture (%)": moisture_values,
    "Bearing Capacity (kN/m²)": bearing_capacity_values,
    "Soil Type": soil_type_values
})

# Save as CSV
soil_data.to_csv("random_soil_data.csv", index=False)
```
